```python
df2['Athlete performance'] = df2['Athlete performance'].str.strip('h')
```

```python
#Difference in speed for the 50k, 50mi male to female
```

```python
df3.groupby(['athlete_gender','race_length'],as_index = False)['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50m race (20 + races min)
```

```python
df3.loc[df3['race_length']=='50mi'].groupby('athlete_age')['athlete_average_speed'] \
.agg(['mean','count']).sort_values('mean', ascending = False).query('count> 19').head(15)
```

```python
#what age groups are the slowest in the 50m race (20 + races min)
```

```python
df3.loc[df3['race_length']=='50mi'].groupby('athlete_age')['athlete_average_speed'] \
.agg(['mean','count']).sort_values('mean', ascending = True).query('count> 19').head(15)
```

```python
#Seasons for the data -> Slower in summer than winter?

#spring 3-5
#Summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
```

```python
df3.head(5)
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x> 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3
```

```python
sns.histplot(data = df3['race_season'])
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values(by = 'mean', ascending = False)
```
